```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
ilkVeri=pd.read_csv('accidents_2005_to_2007.csv',low_memory=False)
ikinciVeri=pd.read_csv('accidents_2009_to_2011.csv',low_memory=False)
ucuncuVeri=pd.read_csv('accidents_2012_to_2014.csv',low_memory=False)
toplamKaza = pd.concat([ilkVeri, ikinciVeri,ucuncuVeri], axis=0)
```

```python
import datetime as dt
import warnings
warnings.filterwarnings('ignore')
selRows = toplamKaza[toplamKaza['Time'].isna()].index
toplamKaza = toplamKaza.drop(selRows, axis=0)
toplamKaza["Time"] = pd.to_datetime(toplamKaza["Time"], format='%H:%M')
toplamKaza["Hour"] = toplamKaza["Time"].dt.hour
toplamKaza["Month"] = pd.DatetimeIndex(toplamKaza["Date"]).month
```

```python
toplamKaza.Hour = toplamKaza.Hour.astype(str)
toplamKaza.Month = toplamKaza.Month.astype(str)
```

```python
f, axs = plt.subplots(3,2, figsize = (14,14), constrained_layout = True)

roadSayisi = toplamKaza.groupby("Road_Type")
toplamRoad=pd.DataFrame(roadSayisi.size().sort_values(ascending=False),columns=["Road_toplam"])
lightSayisi = toplamKaza.groupby("Light_Conditions")
toplamLight=pd.DataFrame(lightSayisi.size().sort_values(ascending=False),columns=["Light_toplam"])
weatherSayisi = toplamKaza.groupby("Weather_Conditions")
toplamWeather=pd.DataFrame(weatherSayisi.size().sort_values(ascending=False),columns=["Weather_toplam"])
surfaceSayisi = toplamKaza.groupby("Road_Surface_Conditions")
toplamSurface=pd.DataFrame(surfaceSayisi.size().sort_values(ascending=False),columns=["road_surface_toplam"])
saatSayisi = toplamKaza.groupby("Hour")
toplamSaat=pd.DataFrame(saatSayisi.size().sort_values(ascending=False),columns=["saat_toplam"])
aySayisi = toplamKaza.groupby("Month")
toplamAy=pd.DataFrame(aySayisi.size().sort_values(ascending=False),columns=["ay_toplam"])
                       
                       

ax=sns.barplot(x=toplamRoad.index, y=toplamRoad["Road_toplam"], data=toplamRoad, ax = axs[0, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplamLight.index,y=toplamLight["Light_toplam"], data=toplamLight, ax = axs[0, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplamWeather.index,y=toplamWeather["Weather_toplam"], data=toplamWeather, ax = axs[1, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplamSurface.index,y=toplamSurface["road_surface_toplam"], data=toplamSurface, ax = axs[1, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplamSaat.index,y=toplamSaat["saat_toplam"], data = toplamSaat, ax = axs[2, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplamAy.index,y=toplamAy["ay_toplam"], data=toplamAy, ax = axs[2, 1])
ax.tick_params(axis = 'x', rotation = 45)
axs[0, 0].title.set_text('Yol Tipine GÃ¶re Kaza OranlarÄ±')
axs[0, 1].title.set_text('IÅŸÄ±k KoÅŸullarÄ±na GÃ¶re Kaza OranlarÄ±')
axs[1, 0].title.set_text('Hava Durumuna GÃ¶re Kaza OranlarÄ±')
axs[1, 1].title.set_text('Yol YÃ¼zey KoÅŸullarÄ±na GÃ¶re Kaza OranlarÄ±')
axs[2, 0].title.set_text('Saate GÃ¶re Kaza OranlarÄ±')
axs[2, 1].title.set_text('Aylara GÃ¶re Kaza OranlarÄ±')
```
